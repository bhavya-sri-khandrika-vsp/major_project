```python
def madev(d, axis=None):
    """ Mean absolute deviation of a signal """
    return np.mean(np.absolute(d - np.mean(d, axis)), axis)

def wavelet_denoising(x, wavelet='db4', level=3):
    coeff = pywt.wavedec(x, wavelet, mode="per")
    sigma = (1/0.6745) * madev(coeff[-level])
    uthresh = sigma * np.sqrt(2 * np.log(len(x)))
    coeff[1:] = (pywt.threshold(i, value=uthresh, mode='hard') for i in coeff[1:])
    return pywt.waverec(coeff, wavelet, mode='per')
```

```python
for wav in pywt.wavelist():
    try:
        filtered = wavelet_denoising(signal, wavelet=wav, level=3)
#         plt.plot(filtered,label='Filtered', color='orange', linewidth = 2)
    except:
        pass
```

```python
def spike_detection(filtered_signal, SPIKE_AMPLITUTE = 1000, SPIKE_GAPE = 100):
    new_filter = [0 for i in range(len(filtered_signal))]
    spike_count = 0
    i = 0
    while i < len(filtered):
        data = filtered[i]
        if data > SPIKE_AMPLITUTE or data < -SPIKE_AMPLITUTE:
            if i+1 < len(filtered):
                if data >=0 and data < filtered[i+1]:
                    # not a peak
                    pass
                elif data <=0 and data > filtered[i+1]:
                    #not a peak
                    pass
                else:
                    spike_count+=1
                    new_filter[i] = data
                    i = i + SPIKE_GAPE
        i+=1
    return new_filter, spike_count
```

```python
def spike_gap_measurment(pure_spike_signal):
    
    spike_gape_dict = {}
    last_spike_position = 0
    spike_count = 0
    
    for i, data in enumerate(pure_spike_signal):
        if data!=0:
            spike_gape_dict[spike_count] = i - last_spike_position
            last_spike_position = i
            spike_count+=1
            
    return spike_gape_dict
```

```python
new_filtered, spike_count = spike_detection(filtered)
```

```python
spike_gap_measurment(new_filtered)
```

# calculating the signal to noise ratio

```python
import numpy as np   
def signaltonoise(filtered, axis=0, ddof=0):
    filtered = np.asanyarray(filtered)
    me = filtered.mean(axis)
    sd = filtered.std(axis=axis, ddof=ddof)
    return np.where(sd == 0, 0, me/sd)
print(signaltonoise(filtered,axis=0,ddof=0))
```

#### Here we got negative SNR, therefore, it means the signal power is low compared to the noise
